```python
import pandas as pd
```

```python
df = pd.read_csv('data/iris.csv')
df.head(10)
```

```python
# data Frame 의 새로운 기능 -> .info() / .describe()
df.info()
```

```python
# .describe() : 평균, 최소값,최대,중위값 등등
df.describe()
```

```python
# 종 분포 그림 그리기
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

```python
sns.catplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', aspect=2)
# 꽃받침(Sepal~)의 길이(x,y)에 따른 종(hue) 분포
```

```python
sns.catplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species', aspect=2)
# 꽃잎의 길이에 따른 종 분포
```

```python
facet = sns.FacetGrid(df, hue='Species',aspect=2)
facet.map(sns.kdeplot,'SepalLengthCm')
facet.add_legend()
plt.show()
# 꽃받침의 길이에 대한 종 밀집도 - facetGrid 그래프로 표현

# facet Grid가 뭐지???
```

```python
# box plot - 종 별로 꽃받침 길이 분포
sns.catplot(kind='box',data=df,x='Species',y='SepalLengthCm',aspect=2)
```

```python
# violin plot - x:종, y:꽃잎 길이
sns.catplot(kind='violin',data=df,x='Species',y='PetalLengthCm',aspect=2)
```

```python
# violin plot - x:종, y:꽃잎 너비
sns.catplot(kind='violin',data=df,x='Species',y='PetalWidthCm',aspect=2)
```

```python
# 꽃잎, 꽃받침의 길이, 너비에 따른 종분포를 pair plot으로...
sns.pairplot(data=df, hue='Species',
            x_vars=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm'],
            y_vars=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm','PetalWidthCm'], height=3)
```
