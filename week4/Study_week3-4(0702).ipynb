```python
# ???
# 세로축이 뭔지 질문
# week3 복습 문제 끝
```

```python
# 1. Sex, Embarked 를 모두 숫자화 + 빈칸 채우기

# Sex
df.loc[df['Sex']=='male','Sex']=0  # 남자면 0
df.loc[df['Sex']=='female','Sex']=1  # 여자면 1

# Embarked
df['Embarked'] = df['Embarked'].fillna("S") # S가 가장 많음
df.loc[df['Embarked']=='C','Embarked']=0  
df.loc[df['Embarked']=='Q','Embarked']=1  
df.loc[df['Embarked']=='S','Embarked']=2  
```

```python
# 2. Fare, Sex, Embarked 이 세 개로 Decision Tree Classifier 모델 만들어
# 정확도 구하고, predict 하기

train = df[['Survived', 'Fare', 'Sex','Embarked']]
x_train = train[[ 'Fare', 'Sex','Embarked']]
y_train = train['Survived']

# x_train
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(x_train, y_train)

# (1)
tree.score(x_train, y_train) # 이 디시전트리의 정확도(황금계수) 

# (2)
x_test = train[[ 'Fare', 'Sex','Embarked']]
pr = tree.predict(x_test)
pr
```

### WEEK4 복습문제
1년 간의 과거 날씨 데이터를 사용하여 도시 내일의 최고 온도 예측하기.  
이 데이터에는 1년 간의 역사적 최대 온도, 지난 이틀 동안의 온도 및 날씨에 관한 정보가 포함되어 있습니다.

위 데이터의 이틀 전 온도, 하루 전 온도, 역사적 평균 최고 온도를 바탕으로 실제 온도를 예측하
는 RandomForestClassifier 모델을 만들어주세요!

***temp_2 , temp_1 , average 데이터 사용**

```python
temp = pd.read_csv('data/temps.csv')

# 필요한 데이터만 추출
sample = temp[['temp_2','temp_1','average']]

inputdata = sample[['temp_2','temp_1']]
targetdata = sample['average']

inputdata
```

```python
# train_test_split 
# => input data를 랜덤으로 균일하게 75%(=x_train) / 25%(=x_test)로 나눠줌

from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(inputdata, targetdata)

x_train # input data의 75%
```

```python
# RandomForest 만들기

from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier()
forest.fit(x_train, y_train) # 75% 데이터로 forest 만듦

# 두 집단 정확도 비교
print('Training set accuracy : ',forest.score(x_train, y_train))
print('test set accuracy : ',forest.score(x_test, y_test))

# 예측
prediction_by_forest = forest.predict(x_test)
prediction_by_forest


# 왜 에러나는거지 fit() 에서.... 내것도 같은 에러 형변환 필요?
```
