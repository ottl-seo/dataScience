### 축구화 판매 회사의 고객분석팀에서 우수 고객을 분류하는 모델 만들기

```python
import pandas as pd

df = pd.read_csv('data/my_customers.csv')
df
```

```python
# 머신러닝 모델은 문자로 된 데이터를 이해할 수 없으므로 숫자 데이터로 바꿔준다
#df.loc[조건, 열] = 바꿀 값

#성별
df.loc[df['성별']=='male', '성별'] = 0 # male -> 0
df.loc[df['성별']=='female', '성별'] = 1 #female -> 1

#지역
df.loc[df['지역']=='서울', '지역'] = 0
df.loc[df['지역']=='경기', '지역'] = 1
df.loc[df['지역']=='제주', '지역']= 2

#관심사
df.loc[df['관심사']=='축구', '관심사'] = 0
df.loc[df['관심사']=='esports', '관심사'] = 1
df.loc[df['관심사']=='야구', '관심사'] = 2

df
```

```python
x_train = df[['나이','성별','지역','관심사']] 
#실제로 우수고객 판단에 영향을 미칠 만한 데이터만 input data로 지정

y_train = df['우수고객분류'] # target data
```

```python
from sklearn.ensemble import RandomForestClassifier

# 앙상블 진행
forest = RandomForestClassifier(n_estimators=10) #트리 개수는 10
forest.fit(x_train, y_train)

print('train set accuracy: ',forest.score(x_train, y_train))
```

## 학습된 forest를 실제 모델에 적용해보기

```python
df_test = pd.read_csv('data/new_customers.csv')
df_test
```

```python
#성별
df_test.loc[df_test['성별']=='male', '성별'] = 0 # male -> 0
df_test.loc[df_test['성별']=='female', '성별'] = 1 #female -> 1

#지역
df_test.loc[df_test['지역']=='서울', '지역'] = 0
df_test.loc[df_test['지역']=='경기', '지역'] = 1
df_test.loc[df_test['지역']=='제주', '지역']= 2

#관심사
df_test.loc[df_test['관심사']=='축구', '관심사'] = 0
df_test.loc[df_test['관심사']=='esports', '관심사'] = 1
df_test.loc[df_test['관심사']=='야구', '관심사'] = 2

df_test
```

```python
x_test = df_test[['나이','성별','지역','관심사']]

prediction = forest.predict(x_test)
prediction
```
