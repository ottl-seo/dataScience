# **판다스로 csv 데이터 불러오기**

```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.csv')

# train #전체 다

train.head() #상단 5개만

# train.head(6) #6개!
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# 산포도 보기
sns.relplot(data=train, x='Age', y='Survived')
```

```python
sns.relplot(data=train, x='Age', y='Survived', hue='Sex', aspect=3)
```

```python
sns.catplot(data=train, x='Pclass', y='Fare', aspect=4, hue='Sex', jitter=True)
```

```python
sns.pairplot(data=train, hue='Sex',
            x_vars=['Pclass', 'Age', 'Fare'],
            y_vars=['Survived','Age', 'Fare'], height=3)
```

```python
sns.catplot(data=train, x='Pclass', y='Fare', kind='box', hue='Sex')

# Box Plot 이해하기
# 최댓값-최솟값 / 중앙값 / 유효한 숫자 범위를 박스로 표현해줌
```
