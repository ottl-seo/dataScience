```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# train 변수에 csv 읽어와서 저장
train = pd.read_csv('data/train.csv')

train
```

```python
survived = train[train['Survived']==1]["SibSp"].value_counts()
dead = train[train['Survived']==0]['SibSp'].value_counts()
print(dead)

dataframe = pd.DataFrame([survived,dead])
dataframe
# train['SibSp'].value_counts() // sibSp 값에 따라 각각의 값이 나온 횟수를 세준다-배열처럼!
# train[train['Survived']==1]["SibSp"].value_counts() // 생존자 중에서 SibSp도 적용~

```

```python
#dataframe.index = ['Survived','Dead']
dataframe.plot(kind='bar',stacked=True)

# 2개 이상의 범주형 변수를 사용하여 각 클래스별로 빈도나 합계를 막대그래프로 비교하려면 
# 위로 쌓은 막대그래프 (stacked bar chart)를 이용할 수 있다.
```
