```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# train 변수에 csv 읽어와서 저장
train = pd.read_csv('data/train.csv')

train
```

```python
# 생존자를 PClass에 따라 바 그래프로 시각화하세요                    

class1 = train[(train['Survived']==1) & (train['Pclass']==1)].shape[0]
class2 = train[(train['Survived']==1) & (train['Pclass']==2)].shape[0]
class3 = train[(train['Survived']==1) & (train['Pclass']==3)].shape[0]
plt.xlabel('Pclass')
plt.ylabel('Survived people')
plt.bar(['1','2','3'],[class1,class2,class3],width=0.75)
plt.show()
```
