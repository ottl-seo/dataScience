```python
# 필요없는 특징 없애기
houseData = df.drop(['id','date'], axis=1) 
# drop(지울것, axis=1)
# axis가 1일 때= '열(row)'을 가리키는 것 --> axis=1 은 항상 고정이라고 생각하자

# houseData = df[['price','bathrooms', ......]] 선택할 게 너무 많은 경우에 drop 사용
```

## 상관계수와 히트맵
### 발크기와 키 사이의 관련성 문제

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

```python
houseData.corr()
## 데이터 간의 상관관계 보여줌 -pandas의 기능!
```

```python
# 히트맵으로 표현하기
plt.figure(figsize=(15,15))
sns.heatmap(houseData.corr(), annot=True, fmt='.2f', square=True)
plt.show()

# sns.heatmap(데이터, annot, fmt, square)
# annot: 숫자 표시 여부
# fmt: 숫자 얼마나 표현할지
```

```python
# 욕실 수(x)에 따른 집값(y) 분포 - 카테고리plot으로
sns.catplot(data=houseData, x='bathrooms',y='price',aspect=2 )
```

```python
# sqft_living 의 최대값 찾기 
houseData['sqft_living'].max()

# .describe() 도 가능하지만, 하나만 딱 골라서 할 수 있다는 장점!
```

```python
# 집값의 최대,최소,평균값
price_max = houseData['price'].max()
price_min = houseData['price'].min()
price_mean = houseData['price'].mean()

print(price_max)
print(price_min)
print(price_mean)
```
