## 사이킷런으로 선형회귀식 나타내기

### 1. data set 구성하기 - Training/ Validation/ Test

train_data : 크기, 욕실수, 층수, ...  
target_data : 집값  

train_data를 8:2 로 나누고, 8할을 Train set, 2할을 Test set으로 둔다  
Train set을 다시 8:2로 나눠서 8할을 Train set, 2할을 Validation set으로 분할- target_data도 똑같이...

````python
import pandas as pd

df = pd.read_csv('data/house.csv')
df.head()
````

````python
houseData = df.drop(['id','date'], axis=1)
houseData.head()
````

````python
from sklearn.model_selection import train_test_split

train_data = houseData.drop(['price'],axis=1) # 가격 빼고 나머지 열을 train data로 사용
target_data = houseData['price']

# train : test = 8:2
x_train, x_test, y_train, y_test = train_test_split(train_data, target_data, test_size=0.2) #test_size 로 비율 정함

# train을 다시 train : Validation = 8:2
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2) #test_size 로 비율 정함

# 각 데이터의 개수
print('모든 데이터', train_data.shape)
print('train set', x_train.shape, y_train.shape)
print('valid set', x_valid.shape, y_valid.shape)
print('test set', x_test.shape, y_test.shape)

# (21613, 18)에서 18이라는 숫자는 열의 개수--
````

````python
from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(x_train, y_train) #학습 시키기

print('train set score', lr.score(x_train, y_train)) # 이미 한번 공부한 내용이므로 더 높은 숫자 나옴
print('valid set score', lr.score(x_valid, y_valid)) #학습된 데이터로 validation set 을 검증
````

집값 == 연속적인 값  
105000 이라고 예상-> 실제값이 104999 라면, O ? X ?  

( 이전의 카테고리값은 1 이라고 예상 -> 1이면 O, 아니면 X 결정 )  

그래서 오차를 보고 오차가 얼마나 작은가에 O, X 결정  
사이킷런의 기능 이용-
```python
from sklearn.linear_model import LinearRegression  
lr = LinearRegression().fit(~~~)
lr.score()
```
.score() 한 값이 **높으면** "학습 잘 된 것" 이라고 이해

// score 는 상관계수에 기인한 **결정계수**

````python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

model = make_pipeline(PolynomialFeatures(2),
                     LinearRegression())

model.fit(x_train, y_train) 
        #학습 시켜주기 fit()

print('train set score', model.score(x_train, y_train))
print('valid set score', model.score(x_valid, y_valid))
````

1차함수에서 2차함수로 높아지니까,  
정확도가 더 높아진 걸 알 수 있음  
1차의 regression보다   
대부분의 상황에서는 polynomial 이 2일 때 더 높은 값이 나온다  

## 미션 1
### polynomial Features 가 3인 모델 만들기
이 틀을 그대로 복사해서 써도 사이킷런 잘 사용하는 것.  
Polynomial의 값을 바꿔주는 것만 해도 Ok

````python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

model3 = make_pipeline(PolynomialFeatures(3),
                     LinearRegression())

model3.fit(x_train, y_train) 
        #학습 시켜주기 fit()

print('train set score', model3.score(x_train, y_train))
print('valid set score', model3.score(x_valid, y_valid))
````

validation set의 score 음수값 나옴  
3차항이 2차항 보다 표현력이 높은 만큼 더 많은 데이터를 필요로 하기 때문  
집값 추정 문제는 3차항을 요구할 정도로 복잡한 문제는 아닌 것일 수도.  

더 많은 데이터로 학습 시켜야 제대로 된 결과가 나올 것.  
우리는 PolynomialFeatures(2) 로 하는 게 가장 정확
